```python
def api_runner():
    global df
    import json
    import ssl
    import urllib.parse
    import urllib.request
    
    import certifi
    
    params = urllib.parse.urlencode(
        {
            "start": "1",
            "limit": "10",
            "convert": "USD",
        }
    )
    
    request = urllib.request.Request(
        f"https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest?{params}",
        headers={
            "Accept": "application/json",
            "X-CMC_PRO_API_KEY": "cbfc4c61d9f149d39b3f9a4fe0cf6a0f",
        },
    )
    
    context = ssl.create_default_context(cafile=certifi.where())
    
    with urllib.request.urlopen(request, context=context) as response:
        data = json.load(response)
    
    pd.set_option('display.max_columns', None)
    df2 = pd.json_normalize(data['data'])
    df2['Timestamp'] = pd.to_datetime('now')
    df = pd.concat([df, df2], ignore_index=True)
    df.sort_values(by = 'name', ascending =True, inplace = True)

from time import time
from time import sleep

for i in range(288):
    api_runner()
    print(f"Run {i+1}/288 completed")
    sleep(30)  # 300 seconds = 5 minutes
    if (i+1) % 5 == 0:
        import os
        import shutil
        csv_path = r'C:\Users\devan\Downloads\crypto_tracking.csv'
        backup_path = r'C:\Users\devan\Downloads\crypto_tracking_backup.csv'
        df.to_csv(csv_path, index=False)
        if os.path.exists(csv_path):
            shutil.copy(csv_path, backup_path)
            print("Backup created successfully")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
bitcoin = df[df['name'] == 'Bitcoin'].sort_values('Timestamp')
sns.lineplot(
    data=bitcoin,
    x='Timestamp',
    y='quote.USD.price',
    marker='o'
)

plt.title('Bitcoin Price Tracking (24 Hours)')
plt.xlabel('Time')
plt.ylabel('Price (USD)')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
