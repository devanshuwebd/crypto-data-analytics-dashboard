```python
def api_runner():
    global df
    import json
    import ssl
    import urllib.parse
    import urllib.request

    import certifi

    params = urllib.parse.urlencode(
        {
            "start": "1",
            "limit": "10",
            "convert": "USD",
        }
    )

    request = urllib.request.Request(
        f"https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest?{params}",
        headers={
            "Accept": "application/json",
            "X-CMC_PRO_API_KEY": "YOUR_API_KEY", ## API Key required...
        },
    )

    context = ssl.create_default_context(cafile=certifi.where())

    with urllib.request.urlopen(request, context=context) as response:
        data = json.load(response)

    
    pd.set_option('display.max_columns', None)
    df2 = pd.json_normalize(data['data'])
    df2['Timestamp'] = pd.to_datetime('now')
    df = pd.concat([df, df2], ignore_index=True)
    df.sort_values(by = 'name', ascending =True, inplace = True)

from time import time
from time import sleep
import shutil
import os
import pandas as pd

df = pd.DataFrame()

for i in range(288):
    api_runner()
    print(f"Run {i+1}/288 completed")
    sleep(300)  # 300 seconds = 5 minutes
    if (i+1) % 5 == 0:  ## Backup on every 25 minutes so that data will be saved in case of crash or stopping of the program...
        csv_path = r'C:\Users\devan\Downloads\Crypto Data Analytics Dashboard\Data\crypto_tracking.csv'
        backup_path = r'C:\Users\devan\Downloads\Crypto Data Analytics Dashboard\Data\crypto_tracking_backup.csv'
        df.to_csv(csv_path, index=False)
        if os.path.exists(csv_path):
            shutil.copy(csv_path, backup_path)
            print("Backup created successfully")

csv_path = r'C:\Users\devan\Downloads\Crypto Data Analytics Dashboard\Data\crypto_tracking.csv'

df.to_csv(csv_path, index=False)

print("Final CSV saved successfully!")
print(f"Total rows collected: {len(df)}")
```
